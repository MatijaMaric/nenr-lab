```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
```

```python
f = open('BATCH.txt', 'r')
batch_err = [float(x.strip()) for x in f.readlines()]
f.close()
f = open('MINIBATCH.txt', 'r')
minibatch_err = [float(x.strip()) for x in f.readlines()]
f.close()
f = open('STOCHASTIC.txt', 'r')
stochastic_err = [float(x.strip()) for x in f.readlines()]
f.close()
```

```python
figure(figsize=(16,10))
plot(batch_err)
plot(minibatch_err)
plot(stochastic_err)
yscale('log')
legend(['batch', 'minibatch', 'stochastic'])
```
